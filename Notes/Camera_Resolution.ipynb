# Scene Coordinates

We'll start by how manim coordinate system works and what are the scene dimensions.

The height and width can be determined by `config.frame_width` and `config.frame_height`. However, it is not the same as pixels.

```python
config.frame_width, config.frame_height
```

We see that the width is 14.222 and height is 8. These values are chosen, because it gives and width/height ratio of 16/9, which is a common screen resolution.

As for the dimensions of scene, the coordinate center of scenes is in the center, which is at (0,0).

The most left point is (-7.1,0), right is (7.1,0), top is (0,4), and button is (0,-4).

```python
# for setup only
def yellow_frame_annotation(framew, frameh):
    d1 = DoubleArrow(framew * LEFT / 2, framew * RIGHT / 2, buff=0).to_edge(DOWN)
    t1 = Text(str(framew)[:6]).next_to(d1, UP)
    d2 = DoubleArrow(frameh * UP / 2, frameh * DOWN / 2, buff=0).to_edge(LEFT)
    t2= Text(str(frameh)).next_to(d2, RIGHT)
    x=Group(d1,d2,t1,t2).set_color(YELLOW)
    return x

def blue_pixel_annotation(framew, frameh,pixelw, pixelh):
    d1 = DoubleArrow(framew * LEFT / 2, framew * RIGHT / 2, buff=0).to_edge(UP)
    t1 = Text(str(pixelw) + " pixel").next_to(d1, DOWN)
    d2 = DoubleArrow(frameh * UP / 2, frameh * DOWN / 2, buff=0).to_edge(RIGHT)
    t2= Text(str(pixelh) + " pixel").next_to(d2, LEFT)
    x=Group(d1,d2,t1,t2).set_color(BLUE)
    return x

annulus = Annulus(inner_radius =1,outer_radius=2,color=WHITE,  stroke_width=10)
```

Now let's go into some more detail.

## Pixel Ratio

### 16/9

First, we'll try to see what the pixel values are:

```python
%%manim -v WARNING  -s -r 160,90 --disable_caching Example
class Example(Scene):
    def construct(self):
        frame_annotation= yellow_frame_annotation(config.frame_width,config.frame_height)
        pixel_annotation= blue_pixel_annotation(config.frame_width,config.frame_height,config.pixel_width,config.pixel_height)
        self.add(frame_annotation, pixel_annotation, annulus)
```

The pixel values are given in the parameters `-r 160,90`. Note that we are maintaining the 16/9 ratio.

### Pixel Ratio Unequal to 16/9

- When the pixel ratio is lower then 16/9 frame_height padded.

## Changing the `frame_width`

- Increasing the value `config.frame_width` will zoom out the Mobject

- Decreasing the value `config.frame_width` will zoom in the Mobject

## Camera Scene

A scene whose camera can be moved around. Let's see a simple example:

```python
param= "-v WARNING  --progress_bar None  -r  500,300  --disable_caching Example"
```

```python
%%manim $param
class Example(MovingCameraScene):
    def construct(self):
        text = Text("Hello World").set_color(BLUE)
        self.add(text)
        self.camera.frame.save_state()
        self.play(self.camera.frame.animate.set(width=text.width * 1.2))
        self.wait(0.3)
        self.play(Restore(self.camera.frame))
```

The example uses the width setting to zoom in, as we saw earlier.

```python
%%manim $param
class Example(MovingCameraScene):
    def construct(self):
        self.camera.frame.save_state()

        ax = Axes(x_range=[-1, 10], y_range=[-1, 10])
        graph = ax.plot(lambda x: np.sin(x), color=WHITE, x_range=[0, 3 * PI])

        dot_1 = Dot(ax.i2gp(graph.t_min, graph))
        dot_2 = Dot(ax.i2gp(graph.t_max, graph))
        self.add(ax, graph, dot_1, dot_2)

        self.play(self.camera.frame.animate.scale(0.5).move_to(dot_1))
        self.play(self.camera.frame.animate.move_to(dot_2))
        self.play(Restore(self.camera.frame))
        self.wait()
```

This example, on the other hand, uses the scale setting to zoom in.

```python
%%manim $param
class Example(MovingCameraScene):
    def construct(self):
        s = Square(color=BLUE, fill_opacity=0.5).move_to(2 * LEFT)
        t = Triangle(color=YELLOW, fill_opacity=0.5).move_to(2 * RIGHT)
        self.add(s, t)
        self.play(self.camera.frame.animate.move_to(s).set(width=s.width*2))
        self.wait(0.3)
        self.play(self.camera.frame.animate.move_to(t).set(width=t.width*2))

        self.play(self.camera.frame.animate.move_to(ORIGIN).set(width=14))
```
